## Busca de preços

<h5><p align = 'justify'>Realizamos uma filtragem nos compostos do GlassNet e criamos a excel com todos os compostos que contém oxigênio, sejam eles óxidos, fosfatos ou sulfatos. Em seguida, pesquisamos o preço de cada um dos compostos no site Sigma Aldrich e adicionamos à tabela. Após esse processo, excluimos os valores NaN e salvamos a tabela em um arquivo csv de nome 'prices'.</h5></p>

```python
compounds_list = []
for compound in list(df_compounds.columns):
   if 'O' in compound:
      compounds_list.append(compound)
      
df_prices_empty = pd.DataFrame()
df_prices_empty['Oxide Coumpounds'] = compounds_list
df_prices_empty['Price per gram(dolar/gram)'] = [None]*len(compounds_list)
#df_prices_empty.to_excel('prices_with_no_filling.excel',index=False)
```

```python
prices = pd.read_csv("prices.csv", delimiter=",")
prices.head(5)
```

<h5><p align = 'justify'>Para conseguirmos ver a distribuição dos preços dos compostos do SciGlass, nós selecionamos apenas as colunas do <i>dataset</i> que estão presentes em nossa tabela de preços por grama de composto, calculamos os preços para cada vidro, caso a soma da fração de cada composto fosse 1. Para além disso, também realizamos o mesmo processo com o módulo de young e a microdureza para conseguirmos conferir como os máximos e mínimos funcionam dentro do nosso subconjunto.</h5></p>

```python
glassnet_with_same_compounds = df_compounds[prices['Oxide Coumpounds'].tolist()]

lines = len(glassnet_with_same_compounds['Ag2CO3'])

all_prices_glassnet = []
all_young_modules_used = []
all_microhardness_used = []
made_by = []
modules_young = df_property['YoungModulus'].tolist()
hardness = df_property['Microhardness'].tolist()

for i in range(lines):
   print(f'{i} de {lines}',end='\r')
   if sum(glassnet_with_same_compounds.iloc[i]) ==1:
      all_prices_per_gram = np.array(prices['Price per gram(dolar/gram)'].tolist())
      made_by.append(glassnet_with_same_compounds.iloc[i])
      value = sum(np.array(glassnet_with_same_compounds.iloc[i])*all_prices_per_gram)
      all_prices_glassnet.append(value)
      all_young_modules_used.append(modules_young[i])
      all_microhardness_used.append(hardness[i])
      
```

```python
sns.histplot(all_prices_glassnet, kde=True)
plt.xlabel('Preço por grama (dolar/grama)', fontsize=12)
plt.ylabel('Frequência', fontsize=12)
plt.title('Histograma de preços\n', fontsize=16)
plt.show()
```

<h5><p align = 'justify'>Por meio do método describe, obtivemos uma tabela abaixo com informações da distribuição preços dos compostos.  </h5></p>

```python
glass_prices = pd.DataFrame()
glass_prices['Prices'] = all_prices_glassnet
glass_prices['Composition'] = made_by
glass_prices.describe()
```

<h5><p align = 'justify'>O mesmo processo de exibição de histograma e informações de distribuição foi feito com os dados de micro dureza e módulo de Young.</h5></p>

```python
plt.figure(figsize=(8, 6))  # 
sns.histplot(all_young_modules_used, bins=500, kde=True, color='blue', edgecolor='black')

plt.xlim(0, 175)

# Adicionar rótulos e título
plt.xlabel('Young Modulus (GPa)', fontsize=14)
plt.ylabel('Frequência', fontsize=14)
plt.title('Histograma do Módulo de Young', fontsize=16)

plt.grid(True)

plt.show()
```
